2. Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”

```python
def search(text):
    
    #finding search section and sending text
    search_icon = driver.find_element(By.XPATH, '//input[@class = "x1lugfcp x19g9edo x1lq5wgf xgqcy7u x30kzoy x9jhf4c x972fbf xcfux6l x1qhh985 xm0m39n x9f619 x5n08af xl565be x5yr21d x1a2a7pz xyqdw3p x1pi30zi xg8j3zb x1swvt13 x1yc453h xh8yej3 xhtitgo xs3hnx8 x1dbmdqj xoy4bel x7xwk5j"]')
    search_icon.send_keys(text)
    time.sleep(5)
        
    #extracting all serched handle
    search_results = driver.find_elements(By.XPATH, '//div[@class = "x9f619 x78zum5 xdt5ytf x1iyjqo2 x6ikm8r x1odjw0f xocp1fn"]//a')
    names = []
    time.sleep(5)
    
    #extracting username
    for i in search_results :
        if i.text.split('\n')[0] != '#' :
            names.append(i.text.split('\n')[0])             
    time.sleep(5)
        
    #clearing search bar
    driver.find_element(By.XPATH, '//div[@class = "_aawn _9-lv"]').click()
 
    return names
    
```

```python
name_list = search('food')
for i in name_list:
    print(i)
```

```python
def follow_unfollow(profile_name):
    #follow the account
    follow_button = driver.find_element(By.XPATH, '//button[@class = "_acan _acap _acas _aj1-"]')
    time.sleep(3)
    follow_button.click()
    time.sleep(3)

    #checking if we have followed
    follow_Unfollow =  driver.find_element(By.XPATH, '//div[@class = "_aacl _aaco _aacw _aad6 _aade"]')
    if follow_Unfollow.get_attribute('innerHTML') == 'Following':
        print('Following the account '+ profile_name)



    #unfollow the account
    follow_button.click()
    time.sleep(3)
    unfollow = driver.find_elements(By.XPATH, '//div[@class = "x1i10hfl x1qjc9v5 xjbqb8w xjqpnuy xa49m3k xqeqjp1 x2hbi6w x13fuv20 xu3j5b3 x1q0q8m5 x26u7qi x972fbf xcfux6l x1qhh985 xm0m39n x9f619 x1ypdohk xdl72j9 x2lah0s xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r x2lwn1j xeuugli xexx8yu x4uap5 x18d9i69 xkhd6sd x1n2onr6 x16tdsg8 x1hl2dhg xggy1nq x1ja2u2z x1t137rt x1q0g3np x87ps6o x1lku1pv x1a2a7pz x1dm5mii x16mil14 xiojian x1yutycm x1lliihq x193iq5w xh8yej3"]')
    unfollow[4].click()
    time.sleep(3)

    if follow_Unfollow.get_attribute('innerHTML') == 'Follow':
        print('Unfollowed the account '+ profile_name)
        
    #going to home
    home()
    return
```

```python
#opening profile of 'dilsefoodie'
open_profile('dilsefoodie')
time.sleep(5)
#scrolling a liitle bit to the posts
driver.execute_script('window.scrollBy(0,300)')
```

```python
posts_window = driver.find_element(By.XPATH, "//div[@style = 'display: flex; flex-direction: column; padding-bottom: 0px; padding-top: 0px; position: relative;']/div/div/a")
```

```python
#liking 30 posts
def like_posts():
    posts_window.click()   

    #finding the like button
    like_button = wait.until(EC.visibility_of_element_located((By.XPATH, '//span[@class = "_aamw"]/div')))
    like_button.click()

    #finding the next button    
    next_button = wait.until(EC.visibility_of_element_located((By.XPATH, '//button[@class = "_abl-"]')))
    next_button.click()
    count = 1

    #loop for 30 posts
    while count < 30:

        like_button = wait.until(EC.visibility_of_element_located((By.XPATH, '//span[@class = "_aamw"]/div')))
        like_button.click()

        next_button = driver.find_elements(By.XPATH, '//button[@class = "_abl-"]')
        time.sleep(2)
        next_button[1].click()
        count+= 1

    post_close = driver.find_element(By.XPATH, '//div[@class = "x160vmok x10l6tqk x1eu8d0j x1vjfegm"]')
    time.sleep(3)
    post_close.click()
    print('Liked the first 30 posts of dilesefoodie')
    return
```

```python
like_posts()
```

```python
#Unliking 30 posts

unlike_posts():
    posts_window.click()   

    #finding the Unlike button
    Unlike_button = wait.until(EC.visibility_of_element_located((By.XPATH, '//span[@class = "_aamw"]/div')))
    Unlike_button.click()

    #finding the next button    
    next_button = wait.until(EC.visibility_of_element_located((By.XPATH, '//button[@class = "_abl-"]')))
    next_button.click()
    count = 1

    #loop for 30 posts
    while count < 30:

        Unlike_button = wait.until(EC.visibility_of_element_located((By.XPATH, '//span[@class = "_aamw"]/div')))
        Unlike_button.click()

        next_button = driver.find_elements(By.XPATH, '//button[@class = "_abl-"]')
        time.sleep(2)
        next_button[1].click()
        count+= 1

    post_close = driver.find_element(By.XPATH, '//div[@class = "x160vmok x10l6tqk x1eu8d0j x1vjfegm"]')
    time.sleep(3)
    post_close.click()
    print('Uniked the first 30 posts of dilesefoodie')
    return
```
